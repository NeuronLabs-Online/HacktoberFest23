```python
from tensorflow.keras.layers import Dense, Flatten, Input
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
```

```python
imagesize =[299,299]
trainPath=r"/content/drive/MyDrive/covidmodel-Data/Dataset/train"
testPath= r"/content/drive/MyDrive/covidmodel-Data/Dataset/test"
```

```python
train_datagen= ImageDataGenerator(rescale = 1./255,
                                    shear_range = 0.2,
                                    zoom_range = 0.2,
                                    horizontal_flip=True)

test_datagen=ImageDataGenerator (rescale = 1./255)
```

```python
training_set = train_datagen.flow_from_directory('/content/drive/MyDrive/covidmodel-Data/Dataset/train',
                                                    target_size=(299, 299),
                                                    batch_size = 16,
                                                    class_mode= 'categorical')

test_set = test_datagen.flow_from_directory('/content/drive/MyDrive/covidmodel-Data/Dataset/test',
                                                target_size=(299, 299),
                                                batch_size = 16,
                                                class_mode= 'categorical')
```

```python
from tensorflow.keras.applications import Xception
xception = Xception(input_shape=imagesize + [3], weights='imagenet', include_top=False)
```

```python
for layer in xception.layers:
  layer.trainable = True
```

```python
x = Flatten()(xception.output)
```

```python
prediction = Dense(4,activation='softmax')(x)
```

```python
model = Model(inputs=xception.input, outputs=prediction)
```

```python
model.summary()
```

```python
model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
```

```python
import tensorflow
my_callbacks=[
    tensorflow.keras.callbacks.EarlyStopping (patience=6)
]
r = model.fit(
    training_set,
    validation_data=test_set,
    epochs=25,
    steps_per_epoch=len(training_set)//16,  # Adjusted value
    validation_steps=len(test_set)//16,  # Adjusted value
    callbacks=my_callbacks
)
```

```python
from keras.callbacks import EarlyStopping
from keras.callbacks import ModelCheckpoint
es = EarlyStopping(monitor= "val_accuracy" , min_delta= 0.01, patience= 3, verbose=1)
mc = ModelCheckpoint(filepath="bestmodel.h5", monitor="val_accuracy", verbose=1, save_best_only= True)
```
