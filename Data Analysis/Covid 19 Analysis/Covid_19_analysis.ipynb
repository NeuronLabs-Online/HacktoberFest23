# **Data Analysis on Covid-19 data**

**Importing the necessary libraries and creating dataframes with given datasets**

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

data_transformed=pd.read_csv("/content/transformed_data.csv")
data_raw=pd.read_csv("/content/raw_data.csv")
display(data_transformed)
```

**Joining the data we need in the two dataframes into a single dataframe**

```python
code=data_transformed["CODE"].unique().tolist()
country=data_transformed["COUNTRY"].unique().tolist()
hdi=[]
tc=[]
td=[]
sti=[]
population=data_transformed["POP"].unique().tolist()
gdp=[]

for i in country:
    hdi.append((data_transformed.loc[data_transformed["COUNTRY"]==i, "HDI"]).sum()/294)
    tc.append((data_raw.loc[data_raw["location"]==i, "total_cases"]).sum())
    td.append((data_raw.loc[data_raw["location"]==i, "total_deaths"]).sum())
    sti.append((data_transformed.loc[data_transformed["COUNTRY"]==i, "STI"]).sum()/294)
    population.append((data_raw.loc[data_raw["location"]==i, "population"]).sum()/294)

aggregated_data = pd.DataFrame(list(zip(code, country, hdi, tc, td, sti, population)),
                               columns = ["Country Code", "Country", "HDI",
                                          "Total Cases", "Total Deaths",
                                          "Stringency Index", "Population"])
display(aggregated_data.head())
```

**Sorting data_transformed dataframe in descending order by Total Cases**

```python
data_transformed=aggregated_data.sort_values(by=["Total Cases"], ascending=False)
display(data_transformed.head())
```

```python
data_transformed=data_transformed.head(10)
display(data_transformed)
```

**Giving inputs to GDP Before Covid and GDP During Covid in data_transformed dataframe**

```python
data_transformed["GDP Before Covid"]=[65279.53, 8897.49, 2100.75,
                            11497.65, 7027.61, 9946.03,
                            29564.74, 6001.40, 6424.98, 42354.41]
data_transformed["GDP During Covid"]=[63543.58, 6796.84, 1900.71,
                            10126.72, 6126.87, 8346.70,
                            27057.16, 5090.72, 5332.77, 40284.64]
display(data_transformed)
```

**Data Visualization**

```python
figure=px.bar(data_transformed, y='Total Cases', x='Country',
            title="Countries with Highest Covid Cases")
display(figure)
```

```python
figure=px.bar(data_transformed, y='Total Deaths', x='Country',
            title="Countries with Highest Deaths")
figure.show()
```

```python
fig=go.Figure()
fig.add_trace(go.Bar(
    x=data_transformed["Country"],
    y=data_transformed["Total Cases"],
    name='Total Cases',
    marker_color='indianred'
))
fig.add_trace(go.Bar(
    x=data_transformed["Country"],
    y=data_transformed["Total Deaths"],
    name='Total Deaths',
    marker_color='lightsalmon'
))
fig.update_layout(barmode='group', xaxis_tickangle=-45)
fig.show()
```

```python
cases=data_transformed["Total Cases"].sum()
deceased = data_transformed["Total Deaths"].sum()

labels=["Total Cases", "Total Deaths"]
values=[cases, deceased]

fig=px.pie(data_transformed, values=values, names=labels,
             title='Percentage of Total Cases and Deaths', hole=0.5)
fig.show()
```

```python
death_rate=(data_transformed["Total Deaths"].sum()/data_transformed["Total Cases"].sum())*100
print("Death Rate=", death_rate)
```
